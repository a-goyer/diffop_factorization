```python
prec = 1000
C = ComplexBallField(prec)
r2, r3 = C(sqrt(2)), C(sqrt(3))
mat = jordan_block(r2, 2).block_sum(matrix([r3]))
ran = MatrixSpace(CC, 3).random_element().change_ring(C)
mat = ~ran * mat * ran
min(c.accuracy() for c in mat.list())
```

```python
chi = mat.charpoly()
eigvals = chi.roots(multiplicities=False, proof=False)
min(c.accuracy() for c in eigvals)
```

```python
from diffop_factorization.polynomials import GCD
d = GCD(chi, chi.derivative())
free_square = chi.quo_rem(d)[0]
eigvals = free_square.roots(multiplicities=False)
min(c.accuracy() for c in eigvals)
```

```python
-log((eigvals[0]-r2).above_abs(), 2)
```

```python
from diffop_factorization.linear_algebra import gen_eigenspaces
GenEigSpaces = gen_eigenspaces(mat)
GenEigSpaces
```

```python
-log((GenEigSpaces[0]['eigenvalue']-r2).above_abs(), 2)
```

```python
GenEigSpaces = gen_eigenspaces(mat, projections=True)
T = matrix(GenEigSpaces[0]['basis'] + GenEigSpaces[1]['basis']).transpose()
pol = GenEigSpaces[0]['projection']
P = ~T * pol(mat) * T; P
```

```python
from diffop_factorization.useful_functions import customized_accuracy, overlaps
M = P - matrix(C, [[1,0,0], [0,1,0], [0,0,0]])
overlaps(M, 0*M), customized_accuracy(M)
```
